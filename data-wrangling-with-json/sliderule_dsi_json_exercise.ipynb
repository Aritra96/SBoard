## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

#### 2. Top 10 major project themes

```python
# 2. Top 10 major project themes

wb_df.mjtheme_namecode.head(10)
```

```python
# Let's create a new df for this 

json_file = json.load(open('data/world_bank_projects.json'))

wb_df2 = json_normalize(json_file, 'mjtheme_namecode')

wb_df2.head(10)
```

```python
# Top 10 project themes

wb_df2.name.value_counts().nlargest(10)
```

```python
# Plotting the bar graph for top 10 major themes

plt.figure(figsize = (10, 6))

wb_df2.name.value_counts().nlargest(10).plot(kind = 'bar')
plt.title('Top 10 major project themes')
plt.annotate('Our mystery theme?', xy = (6, 120),
             xytext = (9, 150),
             arrowprops = dict(facecolor = 'black'),
             ha = 'right', size = 16);
```

The top 10 project themes are:

1. Environment and natural resources management
2. Rural development
3. Human development
4. Public sector governance
5. Social protection and risk management
6. Financial and private sector development
7. < blank >
8. Social dev/gender/inclusion
9. Trade and integration
10. Urban development

The 7th largest project theme seems to be missing. Let us address this in the next task.

#### 3. Creating a dataframe with the missing names filled in

As we might have observed by now, a lot of codes are not mapped with the names of the project which is what has created a separate BLANK category. Our objective here would be to ensure all the codes are mapped accurately with the project name.

```python
# Let us build a dictionary which has all the code-name mappings

dict_map = {}
```

```python
# Iterating through the entire dataframe 

for i in wb_df2.itertuples():
    if i[2] != '':
        dict_map[i[1]] = i[2]
```

```python
# let's see how mapped dictionary now

dict_map
```

```python
# Now we iterate over the named tuples again
# We check for rows where name is blank
# Then we replace it with the corresponding name from our dict_map using .at method

for i in wb_df2.itertuples():
    if i[2] == '':
        wb_df2.set_value(i[0], 'name', dict_map[i[1]])
```

```python
# Result:

wb_df2.name.value_counts()
```

```python
# A more proper answer to the 2nd question:

wb_df2.name.value_counts().nlargest(10)
```

```python
# Plotting a bar graph for correct top 10 major themes

plt.figure(figsize = (10, 6))

wb_df2.name.value_counts().nlargest(10).plot(kind = 'bar')
plt.title('Top 10 major project themes')
plt.annotate('New theme added to our list', xy = (9, 40),
             xytext = (9, 180),
             arrowprops = dict(facecolor = 'black'),
             ha = 'right', size = 16);
```

Therefore, the top 10 project themes are:

1. Environment and natural resources management

3. Human development
4. Public sector governance
5. Social protection and risk management
6. Financial and private sector development
7. Social dev/gender/inclusion
